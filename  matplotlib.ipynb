### 2.5 饼状图
* 饼状图显示一个数据系列中各项的大小于各项总和的比例.
* 饼状图中的数据点显示为整个病症图的百分比
* 如前十大品牌占市场份额图
<p><p>
* 函数画饼状图--注意plt.axes(aspect=1)# 设置x,y轴坐标 1比1,否则画出来的饼状图不是正圆.
<p>matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, hold=None, data=None)
<p>关键参数:x:数据值.autopct:图形上是否显示比例值.explode:模块离圆心的距离



```python
# 饼状图示例
import matplotlib.pyplot as plt

labels=['A','B','C','D']
fracs=[15,30,45,10]

explodes=[0,0.08,0,0]

plt.axes(aspect=1)# 设置x,y轴坐标 1比1,否则画出来的饼状图不是正圆.

# autopct 显示 每个格式中占比, explode控制模块离圆中心的距离
plt.pie(x=fracs,labels=labels,autopct="%0.f%%",explode=explodes,shadow=True)# autopct 显示 每个格式中占比, explode控制模块离圆中心的距离
plt.show()
```

### 2.6 箱形图
* 箱形图,又称为盒须图、盒式图或箱线图
* 是一种用作显示一组数据分散情况资料的统计图
* 上边缘,上四分位数,中位数,下四分位数据,下边缘,异常值
* 函数画箱形状图:  matplotlib.pyplot.boxplot
<p>https://matplotlib.org/api/_as_gen/matplotlib.pyplot.boxplot.html?highlight=boxplot#matplotlib.pyplot.boxplot

```python
#箱形图
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(100)
data=np.random.normal(size=(100,4),loc=0,scale=1)#创建4组数据 每组1000个数据
labels=['A','B','C','D']

plt.boxplot(data,sym='<',whis=1.5,labels=labels)# sym 控制异常值的显示形状,whis:控制 箱子顶部离上线边缘的距离,越大离的越远.

plt.show()
```

## 4 在同一张图上画多个子图(Axes绘图布局方法)
* 绘图布局的创建主要有三种方法：
    <p>在学习这三种方法之前我们首先需要了解以下几点知识点：无论使用何种方法创建，每一个axes都是一个独立的图层,每个图层除了绘图区以外都是透明的
,图层之间相互独立互不影响.当两个axes绘图区之间有重叠的时候，创建命令在前的axes图层位置更靠近底层.axes.set_zorder()方法可改变当前图层层次高度，参数值越小位置越靠近底层

    1、axes = figure.add_axes([left,bottom,width,height])
    <p>add_axes()方法是三种方法中最繁琐的一种方法，因为使用该方法布局需要先计算好绘图区的位置，即参数列表[left,bottom,width,height]的数值。其中

    left：绘图区左侧边缘线与Figure画布左侧边缘线的距离

    bottom：绘图区底部边缘线与Figure画布底部边缘线的距离

    width：绘图区的宽度

    height：绘图区的高度

    它们的单位也不是传统的厘米、毫米这样的长度距离单位，而是Figure画布的百分比，因此它们的取值范围在[0,1]区间内，其中left与width的单位为Figure画布宽度的百分比，bottom与height的单位为Figure画布高度的百分比。
创建区域位置参数及间隔参数可通过fig.subplots_adjust(left,right,top,bottom,wspace,hspace)方法改变。

    2、fig, axes = pyplot.subplots(行数,列数)
    
    <p>pyplot.subplots()方法是三种方法中创建阵列布局最便捷的一种，该方法只需给定行数n和列数m，就可以直接创建出n行m列的绘图区。

    通过pyplot.subplots()方法创建Axes绘图区时，无论是创建单独的一个绘图区还是阵列式的多个绘图区，创建区域在Figure中的位置是固定的，绘图区之间的间隔也是固定的。创建区域位置参数及间隔参数可通过fig.subplots_adjust(left,right,top,bottom,wspace,hspace)方法改变。

    3、axes = figure.add_subplot(行数,列数,序号)
   <p> figure.add_subplot()方法与pyplot.subplots()方法一样，首先在Figure画布上划定出一块区域，然后按行和列进行布局，与pyplot.subplots()方法不同的是该方法一行命令只能创建一个Axes绘图区，即序号参数所指定的那个位置的Axes。


 

```python
#子图示例--figure.add_subplot()
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(1,100)
fig=plt.figure()
ax1=fig.add_subplot(221)
ax2=fig.add_subplot(222)
ax1.plot(x,x)
ax2.bar(x,x)
plt.show()
```

# 带条件的填充

```python
#带条件的填充

import numpy as np
import matplotlib.pyplot as plt
            
x=np.linspace(0,5*np.pi,1000)
y1=np.sin(x)
y2=np.sin(2*x)
            
fig=plt.figure()
ax=plt.gca()
ax.plot(x,y1,color='r')
ax.plot(x,y2,color='b')
ax.fill_between(x,y1,y2,facecolor='yellow')#填充两条线之间的颜色填充为蓝色
ax.fill_between(x,y1,y2,where=y1>y2,facecolor='yellow')#按条件填充,当y1大于y2时,填充黄色
ax.fill_between(x,y1,y2,where=y2>y1,facecolor='green',interpolate=True) # 当y1和y2 交叉值时,自动填充空白处
plt.show()
            
```

# 绘制积分图
        思路:1、绘制一条曲线 2、绘制一个填充为阴影的长方形 跟曲线交接 3、使用fill函数填充颜色  4、写上函数公式

```python
# 绘制积分图

import numpy as np
import matplotlib.pyplot as plt

# plt.style.use('')     
def func(x):
    return -(x-2)*(x-8)+40
x=np.linspace(0,10) #生成x
y=func(x)
     
fig,ax=plt.subplots()
plt.plot(x,y,'r',linewidth=2)
     
a=2
b=9
     
ax.set_xticks([a,b]) #设置x轴坐标
ax.set_yticks([])#设置y轴坐标为空
ax.set_xticklabels([r'$a$',r'$b$']) #设置x轴 显示的标签值
     
plt.figtext(0.9,0.05,'$x$') #把文字画到坐标轴外面
plt.figtext(0.1,0.9,'$y$') #把文字画到坐标轴外面
ix=np.linspace(a,b) # 取 a到b的一些点
iy=func(ix) #设置y轴
ixy=zip(ix,iy) #拼接坐标对
     
verts=[(a,0)]+list(ixy)+[(b,0)]
     
poly=plt.Polygon(verts,facecolor='0.9',edgecolor='0.1')
ax.add_patch(poly)
     
x_math=(a+b)*0.5*0.9
y_math=35
     
plt.text(x_math,y_math,r'$\int_a^b (-(x-2)*(x-8)+40)dx$',fontsize=11,horizontalalignment='center')#画公式,按照中心对其
     
     
plt.ylim(ymin=25)
plt.show()
```

# 绘制多张图
        思路:方法一: 1、定义画布: fig=plt.figure()
               2、定义多个绘制多个绘图区axes:
               ax1=plt.axes(left1, bottom1, width1, height1)
               ax2=plt.axes(left2, bottom2, width2, height2)
               
               3、在每个画布上绘制各自的图形  
                  ax1.scatter(x,y)   
                  ax2.bar(x,y)
                  plt.show()

```python

import numpy as np
import matplotlib.pyplot as plt
    
plt.style.use('ggplot')
x=np.random.randn(200)
y=x+np.random.randn(200)*0.5
 
#设置axes的位置占比    
margin_border=0.1
width=0.6
hight=0.2
margin_between=0.02
    
#定义左边主图 坐标相关位置
left_s=margin_border
bottom_s=margin_border
height_s=width
width_s=width
    
#定义上面图形坐标位置
left_x=margin_border
bottom_x=margin_border+width+margin_between
height_x=hight
width_x=width
    
#定义最右边图形坐标位置
left_y=bottom_x
bottom_y=margin_border
height_y=width
width_y=hight
    
plt.figure(1,figsize=(8,8))
rect_s=[left_s,bottom_s,width_s,height_s]#定义左边axes位置
rect_x=[left_x,bottom_x,width_x,height_x]#定义上边axes位置
rect_y=[left_y,bottom_y,width_y,height_y]#定义右边axes位置
    
axScatter=plt.axes(rect_s)#绘制主图位置
axHisX=plt.axes(rect_x) #绘制上面图位置
axHisY=plt.axes(rect_y) #绘制右面图位置
    
axHisX.set_xticks([])#去掉上面图的x坐标值
axHisY.set_yticks([])#去掉右边图的y坐标值
    
axScatter.scatter(x,y)
    
bin_width=0.25
xymax=np.max([np.max(np.abs(x)),np.max(np.abs(y))]) #求值的范围距离中间的值
lim=int(xymax/bin_width+1)*bin_width #求bins的个数的总宽度
axScatter.set_xlim(-lim,lim)
axScatter.set_ylim(-lim,lim)
    
bins=np.arange(-lim,lim,bin_width)
axHisX.hist(x,bins=bins)#绘制上图
axHisY.hist(y,bins=bins,orientation='horizontal')#绘制右图
    
axHisX.set_xlim(axScatter.get_xlim())
axHisY.set_ylim(axScatter.get_ylim())
    
axHisX.set_title('scatter and hist') #绘制标题

plt.show()
```
